```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis ='columns'), df.target, test_size=0.20)
```

```python
len(X_test)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=41)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot= True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
scores=[ ]
n_estimators=range(1,51)  #example
for i in n_estimators :
    model=RandomForestClassifier(n_estimators=i)
    model.fit(X_train,y_train)
    scores.append(model.score(X_test,y_test))
    print('score:{}, n_estimator:{}'.format(scores[i-1],i))

plt.plot(n_estimators,scores)
plt.xlabel('n_estimators')
plt.ylabel('testing accuracy')
```
